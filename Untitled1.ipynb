```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D
```

```python
def h11(x):
    return 1/(1 + np.exp(-(500*x +30)))

def h12(x):
    return 1/(1 + np.exp(-(500*x - 30)))

def h21(x):
    return h11(x) - h12(x)
```

```python
x1 = np.arange(-1,1,0.01)
X, Y = np.meshgrid(x1,x1)
```

```python
Z = h11(X)
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z)
ax.set_xlabel('x1')
ax.set_ylabel('x1')
ax.set_zlabel('h11')
```

```python
Z = h12(X)
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z)
ax.set_xlabel('x1')
ax.set_ylabel('x1')
ax.set_zlabel('h12')
```

```python
Z = h21(X)
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z)
ax.set_xlabel('x1')
ax.set_ylabel('x1')
ax.set_zlabel('h21')
```
